```python
import pandas
```

```python
database = 'monsters_wikidot.csv'
monsters_df = pandas.read_csv(database).set_index('url', drop=False).fillna("")
monsters_df
```

```python
for index, row in monsters_df.iterrows():
    if row['img'] == "":
        img = input(f"enter the url to use for {row['name']}: ")
        monsters_df.loc[index, 'img'] = img
```

```python
monsters_df.to_csv(database, index=False)
```
