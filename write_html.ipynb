```python
import pandas
```

```python
database = "monsters.csv"
monsters = pandas.read_csv(database).set_index("url")
monsters
```

```python
outfilename = "monsters.html"
outfile = open(outfilename, "w")
```

```python
def convert_monster_to_html(monster):
    result = f'''
        <table style="border: 1px solid black;break-inside: avoid;"><tr>
            <td>{monster["name"]}<br>
                <table>
                    <tr><td>STR {monster["str"]}</td><td>DEX {monster["dex"]}</td><td>CON {monster["con"]}</td></tr>
                    <tr><td>INT {monster["int"]}</td><td>WIS {monster["wis"]}</td><td>CHA {monster["cha"]}</td></tr>
                </table><br>
                {monster["attributes"]}</td>
                <td>{monster["tidbits"]}<br>{monster["descriptions"]}</td>
        </tr></table>
        '''
    return result
```

```python
outfile.write("""
<html><head><title>Animal Forms</title></head>
<body>
""")
for index, row in monsters.iterrows():
    outfile.write(convert_monster_to_html(row))
```

```python
outfile.write("</body></html>")
```

```python
outfile.close()
```
