```python
import pandas
```

```python
disallowed_sources = []
max_CR = 1
excluded_movements = ["fly"]
```

```python
database = "monsters_wikidot.csv"
monsters = pandas.read_csv(database).set_index("url")
source_filter = ~(monsters['source'].isin(disallowed_sources))
CR_filter = monsters['CR'] <= max_CR
movement_filter = True
for movement in excluded_movements:
    movement_filter = (~monsters[movement]) & movement_filter
total_filter = source_filter & CR_filter & movement_filter
monsters = monsters.loc[total_filter]
monsters = monsters.fillna("")
monsters = monsters.sort_values(['CR', 'name'])
monsters
```

```python
outfilename = "monsters_wikidot.html"
outfile = open(outfilename, "w")
```

```python
def convert_monster_to_html(monster):
    image = ""
    CRmap = {0.0:"0", 0.125:"1/8", 0.25:"1/4", 0.5:"1/2"}
    CR = monster['CR']
    CRstr = None
    movement = monster['movement'].replace(' ft.', "'")
    if CR in CRmap.keys():
        CRstr = CRmap[CR]
    elif CR == int(CR):
        CRstr = f"{int(CR)}"
    else:
        CRstr = f"{CR}"
    if monster['img'] != "":
        image = f'<img src="{monster["img"]}" style="max-width: 300px; max-height: 300px"/><br/>'
    result = f'''
        <div style="border: 1px solid black;break-inside: avoid;column-count: 2;">
            {image}<br><h2 style="line-height: 10px;margin-bottom: 10px;margin-top:-10px">{monster["name"]}</h2>
                   
                <table>
                    <tr><td>STR {monster["str"]}</td><td>DEX {monster["dex"]}</td><td>CON {monster["con"]}</td></tr>
                    <tr><td>INT {monster["int"]}</td><td>WIS {monster["wis"]}</td><td>CHA {monster["cha"]}</td></tr>
                </table><br>
                CR: {CRstr}; AC: {monster["AC"]}; HP {monster["HP"]}<br>
                {monster["meta"]}<br>
                {monster["tidbits"]}<br>
                {movement}<br>
                {monster["attributes"]}
        </div>
        '''
    return result
```

```python
outfile.write("""
<html><head><title>Animal Forms</title></head>
<body>
""")
for index, row in monsters.iterrows():
    outfile.write(convert_monster_to_html(row))
```

```python
outfile.write("</body></html>")
```

```python
outfile.close()
```
